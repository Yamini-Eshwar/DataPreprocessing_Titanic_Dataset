### Fill the null values of Age column. 

```python
mean_not_survived = titanic[titanic.Survived == 0 ].Age.mean()
mean_survived = titanic[titanic.Survived == 1 ].Age.mean()
mean_not_survived
mean_survived
```

```python
def fill_age(row):
    if pd.isnull(row['Age']):
        if row["Survived"] == 1:
            return mean_survived
        else:
            return mean_not_survived
    else:
        return row["Age"]
        

titanic['Age'] = titanic.apply(fill_age, axis=1)
```
